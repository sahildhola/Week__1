```python
#Local and global variable in python
gvariable=10
def listsum(numList):


  global gvariable
  gvariable =20
  print(f"value of the gvariable is {gvariable}")
  theSum=0 # this a variable to hold sum
  for i in numList:  # for loop to access all the list members
    theSum=theSum+i  # to sum all the members
  count=len(numList)
  print(f"count of the  list is {count}")
  return theSum,count # return the sum
```

```python
print(f"value of g variable before callin the function {gvariable}")
mylist=[3,14,5,6,17,8,19,0]
Sm,cnt=listsum(mylist)
print(f"sum of the list is {Sm}")
print(f"count of the list is {cnt}")
print(f"value of g variable after callin the function {gvariable}")
```

```python
#use of pandas
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
print(df.shape)
df.head()
```

```python
#iris.data contains only the feature values (like sepal length, petal width, etc.) for each sample.
#iris.target is a separate array that holds the numeric species code (0, 1, or 2) for each sample.
iris = load_iris()
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

```python
import matplotlib.pyplot as plt
plt.hist(df['sepal length (cm)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x='species', y='petal width (cm)', data=df)
plt.title('Petal Width by Species')
plt.show()
```

```python
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='species', data=df)
plt.title('Petal Length vs Petal Width')
plt.show()
```

```python
corr = df.iloc[:, :-1].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```
